```python
import pandas as pd
import numpy as np
students = {'name':["Ali","Faraz","Kamran","Faizan","Nasir","Asad","Danish"],
            'Age':[20,21,24,21,20,23,np.nan],
            'Python':[89,49,78,96,100,78,80],
            'Numpy':[78,87,np.nan,89,75,95,85],
            'Pandas':[40,75,85,48,60,89, 96]}

data = pd.DataFrame(students)

data
```

```python
data.fillna({"Age":data["Age"].mean(), "python":0,"Numpy":0,"pandas":0},inplace= True)
```

```python

data["Total_obt"]=(data["Python"]+data["Numpy"]+data["Pandas"])
data
```

```python
data["Result"]=["pass" if m>150 else "fail" for m in data["Total_obt"]]
data
```

```python
data["percentage"]=round((data["Total_obt"] * 100) /300 ,2 )
data
```

```python
data["Grade"]=["A+" if percentage>=80  else "A" if percentage>=70 else "B" if percentage>=60 else "C" if percentage >=50 else "fail" for percentage in data["percentage"]]
data
```

```python
data["Result2"]=["pass" if(m>=50 and n>=50 and p>=50) else "fail"
                 for m,n,p in zip(data["Python"],data["Numpy"],data["Pandas"])]
```
