```python
house_data['Gender'].fillna('Firm',inplace=True)
```

```python
Totalgender = house_data['Gender'].count()
Totalgender
```

```python
gendersCount = house_data.Gender.value_counts()
gendersCount
```

```python
genderpercentage = (gendersCount/Totalgender)*100
genderpercentage
```

```python
house_data.Gender.value_counts().plot(kind='pie')
```

```python
males = (house_data.Gender == "M").sum() 
males
```

```python
females = (house_data.Gender == "F").sum() 
females
```

```python
Firm = (house_data.Gender == "Firm").sum() 
Firm
```

```python
proportions = [males,females,Firm]
proportions
```

```python
plt.pie(proportions, labels = ['males','females','Firm'], autopct='%1.1f%%')
plt.axis('equal')
plt.title('Gender proportion')
plt.show()
```

```python
frequency = house_data.State.value_counts()
frequency
```

```python
totalstate = house_data.State.count()
totalstate
```

```python
relativefrequency = (frequency / totalstate )* 100
relativefrequency
```

```python
cumulativefrequency = relativefrequency.cumsum()
cumulativefrequency
```
